## 3. HSV 색상으로 얼굴색만 추출하기
#### 3-1 알려진 피부색 범위를 통한 얼굴 색 추출
> 사람의 피부색 범위(african-american은 색상 범위가 다름 주의!)<br>
lower = np.array([20,48,80], dtype ='uint8')<br>
upper = np.array([20,255,255], dtype = 'uint8')<br>

```python
img = cv2.imread("./datas/stark.png")
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

result = img.copy() # copy를 해야지 깊은 복사(서로 id가 달라짐)가 되는 것이고
                    # result = img 하면 포인터로 같은 값을 가리키게 된다(얕은복사)

print(result.shape)
print(result.shape[0])

# blue 채널만 가져오는 것!
for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        # int를 사용함으로 8bit에서 32bit으로 변경되어 오버플로우 예방
        col = (int(img[r,c,0]) + int(img[r,c,1]) + int(img[r,c,2]))/3
        result[r, c, 0] = col
        result[r, c, 1] = col
        result[r, c, 2] = col

# 흑백이지만 3차원이기에 color표현이 가능한 grey영상
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))


#skin color로 사람 얼굴 검출은 한계가 있어 패턴을 통해 사람 얼굴을 추출하는 것으로 발전
```

```python
### 사람의 피부색 범위(african-american은 색상 범위가 다름 주의!)
lower = np.array([20,48,80], dtype ='uint8')
upper = np.array([20,255,255], dtype = 'uint8')
```

```python
for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        if h[r,c] >= 0 and h[r,c] <= 20:
            result[r, c, 0] = img[r, c, 0]
            result[r, c, 1] = img[r, c, 1]
            result[r, c, 2] = img[r, c, 2]
        else:            # 피부색이 아니면 검정색
            result[r, c, 0] = 0 
            result[r, c, 1] = 0
            result[r, c, 2] = 0

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))

#skin color로 사람 얼굴 검출은 한계가 있어 패턴을 통해 사람 얼굴을 추출하는 것으로 발전
```

```python
img = cv2.imread("./datas/stark.png")
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

result = img.copy() # copy를 해야지 깊은 복사(서로 id가 달라짐)가 되는 것이고
                    # result = img 하면 포인터로 같은 값을 가리키게 된다(얕은복사)

print(result.shape)
print(result.shape[0])
lower = np.array([20,48,80], dtype ='uint8')
upper = np.array([20,255,255], dtype = 'uint8')

mask = img.copy()

for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        bH = h[r,c] >= 0 and h[r,c] <= 20
        bS = s[r,c] >=48 and s[r,c] <= 255
        bV = v[r,c] >=80 and v[r,c] <= 255
        
        if bH and bS and bV:
            result[r, c, :] = img[r, c, :]
            mask[r, c, :] = 255
                       
        else:            # 피부색이 아니면 검정색
            result[r, c, :] = 0
            mask[r, c, :] = 0

all = np.hstack([img, mask, result])
# 흑백이지만 3차원이기에 color표현이 가능한 gray영상
plt.imshow(cv2.cvtColor(all, cv2.COLOR_BGR2RGB))
plt.show()
```

## 4. 크로마키 이미지 추출

* 원본 확인

```python
img = cv2.imread("./datas/captainA.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

* 이미지 추출을 위해 이미지에 대한 rgb 범위를 확인

```python
hvs_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, v, s = cv2.split(hvs_img)

plt.figure(figsize = (20,10))
plt.hist(h.ravel(), bins = 255, range= [0,255], color = 'b')
plt.hist(v.ravel(), bins = 255, range= [0,255], color = 'g')
plt.hist(s.ravel(), bins = 255, range= [0,255], color = 'r')
plt.xlim([0,270])
plt.xticks(np.arange(0,260,10))
plt.show()
```

* 이미지 추출

```python
result = img.copy()
mask = img.copy()

for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        hh = h[r,c]>= 55 and h[r,c]<= 65
        vv = v[r,c]>= 240 and v[r,c]<= 250
        ss = s[r,c]>= 250 and s[r,c]<= 255

        if not(hh and ss and vv):
            result[r, c, :]= img[r,c,:]
            mask[r,c,:] = 255
        else:
            result[r, c, :]= 0
            mask[r,c,:] = 0

all = np.hstack((img, result, mask))
plt.imshow(cv2.cvtColor(all, cv2.COLOR_BGR2RGB))
```
