## 1. 커널 마스크(필터) 개념
* 커널을 이미지에 적용하기 전에 간단한 배열 선언을 통해 이해해보자.
* filter를 선언할 때 unit8과 float으로 지정할 때의 차이까지 같이 확인해보자

#### uint8을 사용했을 때 문제(float형 사용하자)

```python
img = np.array([[1,2,3],
               [200,5,200],
               [7,8,9]
               ], dtype='uint8')

kernel = np.array([[0,0,0],
                   [-1,0,-1],
                   [0,0,0]
                  ])

filtered =cv2.filter2D(img, -1, kernel)
print(filtered)
```

#### float을 사용시 0-255의 범위가 넘는 문제를 해결하기 위해 정규화를 사용하자!!

```python
img = np.array([
      [1,2,3],
    [200,5,200],
    [7,8,9]
], dtype="float")

kernel = np.array([[0,0,0],
                   [1,0,-1],
                   [0,0,0]
                  ])
filtered = cv2.filter2D(img, -1, kernel)
print(filtered)
```

* 수학적 공식을 이용한 정규화

```python
# 정규화의 수학적 공식 사용.
ma = np.max(filtered)
mm = np.min(filtered)
n = (filtered - mm) / (ma-mm) * 255 # -2 ~ 0 -> 0 ~ 255
print(n)

n2 = np.abs(filtered) / np.max(np.abs(filtered)) * 255 # 0-2 => 0 ~ 255
print(n2)
```

* normalize 함수를 이용한 정규화

```python
norm = cv2.normalize(filtered, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
                    #None 은 마스크 x
                    #NORM_MINMAX는 양수일 경우에 min-max 스케일러
norm # -2~0까지 이기에 -2는 가장 안밝은것 0은 가장 밝은 것
```

#### 최종 확인

```python
# 음수면 더해주거나 정규화를 해주어라

src = cv2.imread("./datas/Lena.png", cv2.IMREAD_GRAYSCALE)
src2 = src.astype('int16')

kernel = np.array([[0,1,0],
                  [0,0,0],
                  [0,-1,0]
                  ])

filtered = cv2.filter2D(src2, -1, kernel) + 128
print(np.min(filtered)) #음수값이 표현되어있네 확인 가능, uint8은 0부터 나오지만 정상적으로 결과가 나오지 않는다.
print(np.max(filtered))

filtered2 = cv2.filter2D(src2, -1, kernel)
filtered2 = cv2.normalize(filtered, None, 0, 255, cv2.NORM_MINMAX)

print(np.min(filtered2))
print(np.max(filtered2))

cv2.imwrite("./datas/output3.png", filtered2)
```

## 2. 커널 마스크 작성을 통한 화소 영역 처리 이해
* 화소 영역 처리 : 화소의 원값이나 위치를 바탕으로 화소 값을 변경하는 화소의 점 처리과 달리 해당 입력 화소뿐만 아니라 그 주위의 화소 값도 함께 고려하는 공간 영역 연산 


#### 원본 이미지

```python
img = cv2.imread("./datas/kogi.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### 2.1 Original 마스크(원본 그대로)

```python
kernel = np.array([[0,0,0],
                  [0,1,0],
                  [0,0,0]
                  ])

filtered =cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
```

### 2.2 blur 마스크
* 블러링 마스크는 모든 계수가 전체 합이 1이다.
* 디지털 영상에서 세세한 부분은 화소 값이 극단적인 값에 속함. 이 극단적 값을 제거하는 대표적인 방법이 바로 평준화로, 평균값으로 대체하는 것 
* 블러링 회선 마스크의 계수는 평균을 구하는 데 사용되므로 모두 값이 같음

```python
img = cv2.imread("./datas/kogi.png")
K = 3
kernel = np.ones((K,K), np.float32) / (K*K)
print(kernel)

filtered = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
```

```python
img = cv2.imread("./datas/kogi.png")
K = 11
kernel = np.ones((K,K), np.float32) / (K*K) # 121로 나눴기에 더 작아짐
print(kernel)

filtered = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
```

### 2.3 edge 마스크
* edge : 디지털 영상의 밝기가 낮은 값에서 높은 값으로 또는 높은 값에서 낮은 값으로 변하는 지점
<img src=./datas/edge.png>
* 디지털 영상의 에지: 물체 식별, 위치/모양/크기 등을 인지하고 방향성을 탐지할 수 있는 정보 제공

#### 수평 엣지(좌우 픽셀의 차이를 보는 커널)

```python

kernel = np.array([[1,0,-1],
                  [1,0,-1],
                  [1,0,-1]
                  ])

filtered =cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))

# 좌우 차이의 변화량만 보겠다.
```

```python
#### 수직 엣지(상하 픽셀의 차이를 보는 커널)
```

```python
kernel = np.array([[1,1,1],
                  [0,0,0],
                  [-1,-1,-1]
                  ])

filtered =cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))

# 상하 차이의 변화량만 보겠다.
# -2 가 253으로 나올 수도 있기 때문에 정규화 전에는 비정상적인 결과가 나올 수도 있다.
# 다 더했을때 1이되게끔해야만 한다!
```

## 3. 커널 마스크 함수를 통한 화소 영역 처리 이해

### 3.1 가우시안 필터 사용
* 영상의 세세한 부분을 제거하여 부드럽게 하므로 스무딩(Smoothing) 처리라고도 함(스무딩 처리에 사용되는 대표적인 저역통과 필터)
* 수학적으로 잘 정의된 가우시안 함수에서 나옴\
<img src=./datas/gaussian.png>

```python
img = cv2.imread("./datas/brain.jpeg")
gauss =cv2.GaussianBlur(img,(7,7), 0)

plt.imshow(np.hstack([img, gauss]))
```

```python
img = cv2.imread("./datas/brain.jpeg")
gauss =cv2.GaussianBlur(img,(15,15), 0)

plt.imshow(np.hstack([img, gauss]))
```

```python
img = cv2.imread("./datas/brain.jpeg")
gauss =cv2.GaussianBlur(img,(25,25), 0)

plt.imshow(np.hstack([img, gauss]))
```

### 3.2 소벨 엣지 필터(1차 미분 마스크)

```python
# 엄청난 노이즈로 활용 가치가 없다.
out1 = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)#(CV_64F : 알아서 실수로 만듬)
out2 = cv2.Sobel(gauss, cv2.CV_64F, 1, 0, ksize=3)

plt.imshow(np.hstack([out1, out2]))
```

* 위 이미지는 엣지필터로 바로 사용할수가 없다.
* 소벨 엣지필터로 보면 엄청나게 노이즈가 있기 때문이다.
* 따라서 **미디언 필터로 잡음을 제외하고(블러링을 한 후) edge detection을 하자**

```python
median = cv2.medianBlur(img, 5)

plt.imshow(np.hstack([img, median]))
```
