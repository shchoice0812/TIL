```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from matplotlib import font_manager, rc
#matplotlib 한글 폰트 오류 문제 해결
font_path='./dataset/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
```

## 1. 데이터 살펴보기

```python
df = pd.read_csv('./dataset/sonar.csv', header=None)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## 2. seed 설정

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

## 3. Attribute, Class 나누기

```python
dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]
```

## 4. Categorical Class를 one-hot encoding 변환

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

## 5. Training set, Test set 구분(과적합 방지를 위해, Sample 분할)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

## 6. 딥러닝 모델 설정

```python
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## 7. 모델 컴파일 및 실행

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=200, batch_size=5)
```

## 8. 모델 저장 및 재호출

```python
model.save("my_model.h5")

del model
model = load_model('my_model.h5')
```

## 9. Test_set으로 모델 정확도 검증

```python
print("Accuray : %.4f" % (model.evaluate(X_test, Y_test)[1]))
```
