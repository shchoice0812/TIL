```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

# 영화 소개 웹페이지 URL
value_list = ['133855', '128397', '93004', '135716', '131271']

for value in value_list:
    # 박스오피스 영화제목, 평점 출력
    url = 'https://movie.daum.net/moviedb/main?movieId='+value
    soup = urlopen(url)
    bs = BeautifulSoup(soup, 'html.parser')

    title = bs.find('strong', class_ = 'tit_movie').string
    score = bs.find('em', class_='emph_grade').string
    print("영화 제목 : ", title)
    print("영화 점수 : ", score)

    # 박스오피스 댓글 내용 10개씩 출력
    reviews_list = []

    for n in range(3):
        url = 'https://movie.daum.net/moviedb/grade?movieId='+value+'&type=netizen&page={}'.format(n+1)
        soup = urlopen(url)
        bs = BeautifulSoup(soup, 'html.parser')
        reviews = bs.find_all('p', class_ = 'desc_review')    
    
        for review in reviews:
            reviews_list.append(review.get_text().strip().replace('\n','').replace('\t','').replace('\r',''))

    print("영화 리뷰 :")
    number = 0        
    for review in reviews_list:
        if review is "": continue
        if number == 10 : break
        print("  - ", number+1, "번째 리뷰: ", review)
        number = number + 1
        
    print("-"* 130)
```
